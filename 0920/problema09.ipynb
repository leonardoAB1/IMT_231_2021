Evalúe cada uno de los siguientes bloques de código y determine la complejidad O(·) para los mejores y peores escenarios.

```python
#a) Peor escenario: O(n)
#Mejor escenario: Cuando n sea igual o menor a 0 -> O(1).
def fuctA(n):                           #O(n)
    suma = 0                            #O(1)
    for i in range(n):                  #O(n)*1
        if i % 2 == 0:                  #O(1)*0.5
            suma += i                   #O(1)
#b) Peor escenario: O(logn)
#   Mejor escenario: O(1) Cuando n sea menor a cero
def fuctB(n):                           #O(logn)
    suma = 0                            #O(1)
    i = n                               #O(1)
    while i > 0:                        #O(logn)*1
        suma += i                       #O(1)
        i = i / 2                       #O(1)
        
# c) Peor escenario: O(n^2)
#Mejor escenario:   Cuando n sea igual o menor a 0 -> O(1)
#                   Cuando los valores a iterar sean pares-> 0(1)
def fuctC(n):                           #O(n^2)
    for i in range(n):                  #O(n)*O(n)
        suma = 0                                        #Suma no estaba definida
        if i % 3 == 0:                  #O(n)
            for j in range(int(n / 2)): #O(n)           #Range no acepta floats
                suma += j               #O(1)
        elif i % 2 == 0:                
            for j in range(5):          #O(1)
                suma += j               #O(1)
        else: 
            for j in range(n):          #O(n)
                suma += j               #O(1)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import timeit
from functools import partial

def plotTC(fn, nMin, nMax, nInc, nTests):
    """
    Run timer and plot time complexity
    """
    print('Analyzing Algorithms...')
    x = []
    y = []
    for i in range(nMin, nMax, nInc):
        N = i
        testNTimer = timeit.Timer(partial(fn, N))
        t = testNTimer.timeit(number=nTests)
        x.append(i)
        y.append(t)
    p1 = plt.plot(x, y, 'o')
```

```python
plotTC(fuctA, 10, 10**4, 100, 100)
plt.show()
```

```python
plotTC(fuctB, 10, 10**7, 10000, 100)
```

```python
plotTC(fuctC, 10, 1000, 10, 100)
```
